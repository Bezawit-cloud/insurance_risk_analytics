Outlier Detection

```python
sns.boxplot(x=df['TotalPremium'])
plt.show()
sns.boxplot(x=df['TotalClaims'])
plt.show()
```


The box plots for TotalPremium and TotalClaims both lead to the same short and crucial conclusion about the dataset:

Both variables exhibit extreme right-skewness with a significant number of high-value outliers.

This distribution is characterized by:

- Massive Data Concentration: The vast majority of values for both premiums and claims are tightly clustered at very low amounts (close to zero).

- High-Impact Outliers: A few rare events (outliers) have extremely large values that disproportionately inflate the mean (average) and dominate the total risk.

- Modeling Difficulty: These variables are not normally distributed, requiring statistical data transformation (like a log transform) before they can be effectively used in most predictive models.

The primary conclusion is that the risk and financial exposure in this dataset are driven by the rare, high-severity events (the outliers), not the common, low-value occurrences.

# Claims Over Time (Line Plot)

```python
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'])
monthly = df.groupby("TransactionMonth")["TotalClaims"].sum()
monthly.plot(figsize=(10,4))
plt.title("Monthly Claims Trend")
plt.show()
```

Claims Over Time  Temporal Trend

- 2014: Claims remained relatively stable throughout the year.

- Late 2014  Early 2015: Slight increase in claims observed.

- April 2015: Noticeable spike in claims  could indicate seasonal effect, special events, or reporting anomaly.

- July 2015 onwards: Claims start decreasing again, approaching normal levels.

Interpretation

- The data shows seasonality in claims, with a peak around April 2015.

Understanding these trends can help the insurance company:

- Prepare for higher claim periods

- Adjust reserves and risk models accordingly

- Plan marketing or premium adjustments seasonally
